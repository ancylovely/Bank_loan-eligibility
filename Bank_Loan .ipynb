```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred_log=logmodel.predict(x_test)
```

```python
y_pred_log
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_pred_log,y_test)
```

```python
accuracy_score(y_pred_log,y_test)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
knn.fit(x_train,y_train)
```

```python
y_knn=knn.predict(x_test)
```

```python
confusion_matrix(y_knn,y_test)
```

```python
accuracy_score(y_knn,y_test)
```

```python
# NB
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_gnb=gnb.predict(x_test)
```

```python
confusion_matrix(y_gnb,y_test)
```

```python
accuracy_score(y_gnb,y_test)
```

```python
#svm
from sklearn.svm import SVC
sv=SVC(kernel="rbf")
sv.fit(x_train,y_train)
```

```python
y_rbf=sv.predict(x_test)
```

```python
confusion_matrix(y_rbf,y_test)
```

```python
accuracy_score(y_rbf,y_test)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion="entropy")
dtc.fit(x_train,y_train)
```

```python
y_dtc=dtc.predict(x_test)
```

```python
confusion_matrix(y_dtc,y_test)
```

```python
accuracy_score(y_dtc,y_test)
```
